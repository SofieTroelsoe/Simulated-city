```python
# Top 10 Danish baby names (2024) from Danmarks Statistik
# Source: https://www.dst.dk/en/Statistik/emner/borgere/navne/navne-til-nyfoedte

danish_top10_names = [
    "Emma",
    "Ella",
    "Luna",
    "Alma",
    "Frida",
    "Olivia",
    "Ida",
    "Agnes",
    "Karla",
    "Nora",
]

danish_top10_names
```

```python
# Merge `random_names` and `danish_top10_names` into `random_names`
# Preserve order and remove duplicates while keeping original random_names first
combined = []
seen = set()

# helper to iterate a list safely
for lst_name in ('random_names', 'danish_top10_names'):
    if lst_name in globals() and isinstance(globals()[lst_name], list):
        for n in globals()[lst_name]:
            if n not in seen:
                combined.append(n)
                seen.add(n)

random_names = combined

random_names
```

```python
# Monster energy drink variations ranked by popularity (most to least popular)
monster_variations = [
    "Monster Energy (Original)",
    "Monster Ultra",
    "Monster Zero Ultra",
    "Monster Assault",
    "Monster Pipeline Punch",
    "Monster Rehab",
    "Monster Absolutely Zero",
    "Monster Lo-Carb",
    "Monster Nitro",
    "Monster Juiced",
    "Monster Hydro",
    "Monster Import",
    "Monster Muscle",
    "Monster Mule",
    "Monster Punch",
]

print("Monster Energy Drink Variations (Most to Least Popular):")
for i, flavor in enumerate(monster_variations, 1):
    print(f"{i}. {flavor}")
```

```python
# Assign a random Monster variation to each name in random_names
name_monster_pairs = {}
for name in random_names:
    if name == "Alma":
        name_monster_pairs[name] = "Monster Muscle"
    elif name == "Eva":
        name_monster_pairs[name] = "Monster Pipeline Punch"
    else:
        name_monster_pairs[name] = random.choice(monster_variations)

# Display the assignments
print("Name-Monster Energy Drink Pairings:")
print("-" * 50)
for name, monster in name_monster_pairs.items():
    print(f"{name:15} -> {monster}")
```

```python
# Rank Monster variations by popularity among random_names and show who likes what
from collections import Counter

# Count how many people like each Monster variation
variation_counts = Counter(name_monster_pairs.values())

# Create a dictionary of variation -> list of people who like it
variation_to_people = {}
for name, monster in name_monster_pairs.items():
    if monster not in variation_to_people:
        variation_to_people[monster] = []
    variation_to_people[monster].append(name)

# Sort by count (most popular first)
ranked_variations = sorted(variation_counts.items(), key=lambda x: x[1], reverse=True)

# Display the ranking
print("Monster Variations Ranked by Popularity:")
print("=" * 60)
for rank, (variation, count) in enumerate(ranked_variations, 1):
    people = ", ".join(sorted(variation_to_people[variation]))
    print(f"{rank}. {variation} ({count} {"person" if count == 1 else "people"})")
    print(f"   Likes: {people}")
    print()
```

```python
import matplotlib.pyplot as plt

# Extract variations and counts from ranked list
variations = [v[0] for v in ranked_variations]
counts = [v[1] for v in ranked_variations]

# Create the bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(range(len(variations)), counts, color='steelblue', edgecolor='black')

# Customize the chart
plt.xlabel("Monster Variation", fontsize=12, fontweight='bold')
plt.ylabel("Number of People", fontsize=12, fontweight='bold')
plt.title("Monster Energy Drink Variations Ranked by Popularity", fontsize=14, fontweight='bold')
plt.xticks(range(len(variations)), variations, rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

# Add value labels on top of bars
for i, (bar, count) in enumerate(zip(bars, counts)):
    plt.text(bar.get_x() + bar.get_width()/2, count + 0.05, str(count), 
             ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Assign each person from random_names a random number from 1 to 10
name_number_pairs = {}
for name in random_names:
    name_number_pairs[name] = random.randint(1, 10)

# Display the assignments
print("Name-Random Number Assignments (1-10):")
print("-" * 40)
for name, number in name_number_pairs.items():
    print(f"{name:15} -> {number}")
```

```python
# Calculate total weekly purchases for each Monster variation
# based on how many times each person buys their assigned variation
monster_weekly_totals = {}

for name, monster in name_monster_pairs.items():
    frequency = name_number_pairs[name]
    if monster not in monster_weekly_totals:
        monster_weekly_totals[monster] = 0
    monster_weekly_totals[monster] += frequency

# Sort by total purchases (descending)
sorted_totals = sorted(monster_weekly_totals.items(), key=lambda x: x[1], reverse=True)

# Create the bar chart
plt.figure(figsize=(14, 6))
variations_labels = [v[0] for v in sorted_totals]
purchase_totals = [v[1] for v in sorted_totals]

bars = plt.bar(range(len(variations_labels)), purchase_totals, color='#FF6B35', edgecolor='black', linewidth=1.5)

# Customize the chart
plt.xlabel("Monster Variation", fontsize=12, fontweight='bold')
plt.ylabel("Total Weekly Purchases", fontsize=12, fontweight='bold')
plt.title("Weekly Monster Energy Drink Purchases by Variation", fontsize=14, fontweight='bold')
plt.xticks(range(len(variations_labels)), variations_labels, rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

# Add value labels on top of bars
for bar, total in zip(bars, purchase_totals):
    plt.text(bar.get_x() + bar.get_width()/2, total + 0.2, str(total), 
             ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Also print the sorted totals
print("\nMonster Variations Ranked by Weekly Purchases:")
print("=" * 50)
for rank, (variation, total) in enumerate(sorted_totals, 1):
    print(f"{rank}. {variation}: {total} purchases/week")
```

```python
# Create a pie chart showing weekly Monster purchases by variation
plt.figure(figsize=(12, 8))

# Use the sorted_totals data for the pie chart
pie_labels = [v[0] for v in sorted_totals]
pie_values = [v[1] for v in sorted_totals]

# Create the pie chart with a nice color palette
colors = plt.cm.Set3(range(len(pie_labels)))
wedges, texts, autotexts = plt.pie(pie_values, labels=pie_labels, autopct='%1.1f%%',
                                     colors=colors, startangle=90, textprops={'fontsize': 9})

# Enhance text readability
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontweight('bold')

plt.title("Weekly Monster Energy Drink Purchases Distribution by Variation", 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
class Person:
    """Represent a person with name, age, and home location."""
    def __init__(self, name: str, age: int, latitude: float, longitude: float):
        self.name = name
        self.age = age
        self.latitude = latitude
        self.longitude = longitude
    
    def __repr__(self) -> str:
        return f"Person(name={self.name!r}, age={self.age}, location=({self.latitude}, {self.longitude}))"
    
    def get_location(self) -> tuple:
        """Return home location as (latitude, longitude) tuple."""
        return (self.latitude, self.longitude)

# Create Person instances for each name in random_names with sample ages and Copenhagen area locations
# Copenhagen area coordinates: ~55.6761N, 12.5683E
import random as rnd

people = {}
for name in random_names:
    age = rnd.randint(18, 65)  # Random age between 18 and 65
    # Varying locations around Copenhagen area
    latitude = 55.6761 + rnd.uniform(-0.1, 0.1)
    longitude = 12.5683 + rnd.uniform(-0.1, 0.1)
    people[name] = Person(name, age, latitude, longitude)

# Display all people
print("People with their home locations:")
print("=" * 80)
for name, person in people.items():
    print(f"{person}")
```

```python
import math

# Copenhagen city center coordinates
copenhagen_center_lat = 55.6761
copenhagen_center_lon = 12.5683

# === Row 1: Sort by Name (Alphabetical Order) ===
print("ALPHABETICAL ORDER (by Name):")
print("=" * 80)
sorted_by_name = sorted(people_list, key=lambda p: p['name'])
for person in sorted_by_name:
    print(f"Name: {person['name']:10} | Age: {person['age']:2} | Location: ({person['latitude']:.4f}, {person['longitude']:.4f})")

# === Row 2: Sort by Age Order ===
print("\n\nAGE ORDER (Youngest to Oldest):")
print("=" * 80)
sorted_by_age = sorted(people_list, key=lambda p: p['age'])
for person in sorted_by_age:
    print(f"Name: {person['name']:10} | Age: {person['age']:2} | Location: ({person['latitude']:.4f}, {person['longitude']:.4f})")

# === Row 3: Sort by Distance from Copenhagen Center ===
print("\n\nDISTANCE FROM COPENHAGEN CENTER (Closest to Farthest):")
print("=" * 80)

# Calculate distance for each person
def calculate_distance(person):
    """Calculate distance from Copenhagen center using Euclidean distance."""
    lat_diff = person['latitude'] - copenhagen_center_lat
    lon_diff = person['longitude'] - copenhagen_center_lon
    distance = math.sqrt(lat_diff**2 + lon_diff**2)
    return distance

# Add distance to each person for sorting
people_with_distance = []
for person in people_list:
    person_copy = person.copy()
    person_copy['distance'] = calculate_distance(person)
    people_with_distance.append(person_copy)

# Sort by distance
sorted_by_distance = sorted(people_with_distance, key=lambda p: p['distance'])
for person in sorted_by_distance:
    print(f"Name: {person['name']:10} | Age: {person['age']:2} | Distance: {person['distance']:.4f} | Location: ({person['latitude']:.4f}, {person['longitude']:.4f})")
```

```python
# Create a grayscale map of Copenhagen showing each person's location
# Colors represent ages (19-29)

# Create the figure and axis
fig, ax = plt.subplots(figsize=(16, 14))

# Set grayscale background
ax.set_facecolor('#e8e8e8')

# Get age range for color mapping
min_age = min([p['age'] for p in people_list])
max_age = max([p['age'] for p in people_list])

# Use a colormap to represent ages
cmap = plt.cm.get_cmap('tab20')  # Use tab20 colormap with distinct colors
norm = plt.Normalize(vmin=min_age, vmax=max_age)

# Plot each person
scatter = ax.scatter([], [], s=150, edgecolor='black', linewidth=1.5)  # Dummy scatter for collecting handles

for person in people_list:
    age = person['age']
    # Get color based on age
    color = cmap(norm(age))
    
    # Plot the dot (marker) for the person's location
    ax.scatter(person['longitude'], person['latitude'], 
               s=150, color=color, edgecolor='black', linewidth=1.5, zorder=3, alpha=0.8)
    
    # Add the person's name above the dot in the same color
    ax.text(person['longitude'], person['latitude'] + 0.004, person['name'], 
            fontsize=10, ha='center', va='bottom', color=color, fontweight='bold', zorder=4)

# Add Copenhagen center marker
ax.scatter(12.5683, 55.6761, s=300, marker='*', color='yellow', 
           edgecolor='black', linewidth=2.5, zorder=5, label='Copenhagen Center')

# Customize the map
ax.set_xlabel('Longitude', fontsize=12, fontweight='bold')
ax.set_ylabel('Latitude', fontsize=12, fontweight='bold')
ax.set_title('Copenhagen - People Locations Map (Color = Age)', fontsize=16, fontweight='bold', pad=20)

# Add grid for reference
ax.grid(True, alpha=0.4, linestyle='--', color='gray', linewidth=0.5)

# Set aspect ratio
ax.set_aspect('equal', adjustable='box')

# Add a colorbar to show age-color mapping
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, pad=0.02, label='Age (Years)')
cbar.set_label('Age (Years)', fontsize=12, fontweight='bold')

# Add legend for Copenhagen Center
ax.legend(loc='upper right', fontsize=11)

# Set axis limits with some padding
ax.set_xlim(12.4, 12.75)
ax.set_ylim(55.55, 55.8)

plt.tight_layout()
plt.show()

# Print summary with age color information
print("\nMap Summary:")
print("=" * 50)
print(f"Total people plotted: {len(people_list)}")
print(f"Map center: Copenhagen (55.6761N, 12.5683E)")
print(f"Color scheme: Colors represent ages (19-29 years)")
print(f"Marker: Yellow star = Copenhagen city center")
print(f"\nAge Range: {min_age} - {max_age} years")
print("\nPeople by Age:")
print("-" * 50)
sorted_by_age_for_display = sorted(people_list, key=lambda p: p['age'])
for person in sorted_by_age_for_display:
    print(f"Age {person['age']:2}: {person['name']}")
```
